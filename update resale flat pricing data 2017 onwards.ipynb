```python
import pandas as pd
import numpy as np

data = pd.read_csv('resale_flat_prices_based_on_registration_date_from_jan_2017_onwards.csv')
data['CCR'] = 0
data['RCR'] = 0
data['OCR'] = 0
data
```

```python
unqtowns = data['town'].unique()
print(unqtowns)
print(len(unqtowns))
```

```python
ccr = ['CENTRAL AREA', 'BUKIT TIMAH'] 
rcr = ['BISHAN','BUKIT MERAH','GEYLANG','KALLANG/WHAMPOA','MARINE PARADE','QUEENSTOWN', 'TOA PAYOH']
ocr = ['ANG MO KIO', 'BEDOK', 'BUKIT BATOK', 'BUKIT PANJANG','CHOA CHU KANG', 'CLEMENTI','HOUGANG', 'JURONG EAST', 'JURONG WEST', 'PASIR RIS',
      'PUNGGOL', 'SEMBAWANG', 'SENGKANG', 'SERANGOON', 'TAMPINES', 'WOODLANDS', 'YISHUN']

len(ccr) + len(rcr) + len(ocr)
```

```python
for i in range(len(data)) :
    if data.iloc[i,1] in ccr:
        data.iloc[i,11] = 1
    elif data.iloc[i,1] in rcr:
        data.iloc[i,12] = 1
    elif data.iloc[i,1] in ocr:
        data.iloc[i,13] = 1
        
```

```python
data
```

```python
data['low_storey'] = 0
data['high_storey'] = 0
```

```python
low = ['01 TO 03','04 TO 06', '07 TO 09']

for i in range(len(data)) :
    if data.iloc[i,5] in low:
        data.iloc[i,14] = 1
    elif data.iloc[i,5] not in low:
        data.iloc[i,15] = 1
        
```

```python
data['price per sqm'] = data['resale_price']/data['floor_area_sqm']
data['ln price'] = np.log10(data['resale_price'])
```
